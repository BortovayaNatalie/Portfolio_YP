# Проект для «Викишоп» с BERT

Интернет-магазин «Викишоп» запускает новый сервис. Теперь пользователи могут редактировать и дополнять описания товаров, как в вики-сообществах. То есть клиенты предлагают свои правки и комментируют изменения других. Магазину нужен инструмент, который будет искать токсичные комментарии и отправлять их на модерацию. 

Обучим модель классифицировать комментарии на позитивные и негативные. Заказчиком предоставлен набор данных с разметкой о токсичности правок.

Необходимо потроить модель со значением метрики качества *F1* не меньше 0.75.

Для выполнения проекта применим *BERT*.

## Обучение

### Напишем класс `WorkFlow` для инициализации моделей и хранения результатов их работы. 

```python
class WorkFlow():
    '''Класс для обучения переданных моделей и логирования результатов'''
    
    def __init__(self) -> None:
        '''Инициализация объекта класса'''
        self.results = {}
        return None
        
    def add_model(self, model, name, X, y, X_test, y_test, scoring='f1',
                  fit=True, report='print', vectorizer=None) -> None:
        '''Заводит в поток переданную модель, обучает, предсказывает, сохраняет результаты'''
        self.results[name] = {}
        if vectorizer:
            model = make_pipeline(vectorizer, model)
        if fit:
            model = model.fit(X, y)
        self.results[name]['model'] = model
        predictions = model.predict(X)
        self.results[name]['predictions'] = predictions
        predictions_test = model.predict(X_test)
        self.results[name]['predictions_test'] = predictions_test
        if fit:
            score = cross_val_score(model, X, y, scoring='f1').mean()
            self.results[name][scoring + ' CV score'] = score
        if report == 'print':
            print(name, 'mean', scoring, 'score on CV =', score.round(decimals=3))
        elif report:
            self.report(name, X_test, y_test, test='_test')
        return None
        
    def report(self, name, X, y, test='') -> None:
        '''Выводит на экран отчёт по классификации'''
        cr = classification_report(y, self.results[name]['predictions' + test], output_dict=True, zero_division=0)
        display(pd.DataFrame(cr).round(decimals=3).transpose())
        return None
```

```python
wf = WorkFlow()
```

Подготовим визуализацию матрицы ошибок.

```python
def double_plot(y, name_first, name_second, test='') -> None:
    '''Выводит на экран две матрицы ошибок'''
    cm_l = confusion_matrix(y, wf.results[name_first]['predictions' + test])
    cm_r = confusion_matrix(y, wf.results[name_second]['predictions' + test])

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    cmd_l = ConfusionMatrixDisplay(cm_l)
    cmd_l.plot(ax=ax1)
    ax1.set_title(name_first)
    
    cmd_r = ConfusionMatrixDisplay(cm_r)
    cmd_r.plot(ax=ax2)
    ax2.set_title(name_second)
    
    plt.show()
    return None
```

### Проведем векторизацию двумя способами, в дальнейшем сравним их.

```python
bow = CountVectorizer(stop_words='english', dtype=np.float32)
tfidf = TfidfVectorizer(stop_words='english', dtype=np.float32)
```

### LogisticRegression

```python
%%time
wf.add_model(
    LogisticRegression(random_state=31416, solver='lbfgs', n_jobs=-1, C=5), 'LogisticRegression BOW',
    X, y, X_test, y_test, vectorizer=bow
)
wf.add_model(
    LogisticRegression(random_state=31416, solver='lbfgs', n_jobs=-1, C=5), 'LogisticRegression TFIDF',
    X, y, X_test, y_test, vectorizer=tfidf
)
```

```python
double_plot(y, 'LogisticRegression BOW', 'LogisticRegression TFIDF')
```

### LightGBM

```python
%%time
wf.add_model(
    lgb.LGBMClassifier(random_state=31416, n_jobs=-1), 'LGBM BOW',
    X, y, X_test, y_test, vectorizer=bow
)
wf.add_model(
    lgb.LGBMClassifier(random_state=31416, n_jobs=-1), 'LGBM TFIDF',
    X, y, X_test, y_test, vectorizer=tfidf
)
```

```python
double_plot(y, 'LGBM BOW', 'LGBM TFIDF')
```

### LinearSVС

```python
%%time
wf.add_model(
    LinearSVC(dual=False, random_state=SEED, max_iter=1000, class_weight='balanced', C=0.5), 'LinearSVC BOW',
    X, y, X_test, y_test, vectorizer=bow
)
wf.add_model(
    LinearSVC(dual=False, random_state=SEED, max_iter=1000, class_weight='balanced', C=0.5), 'LinearSVC TFIDF',
    X, y, X_test, y_test, vectorizer=tfidf
)
```

```python
double_plot(y, 'LinearSVC BOW', 'LinearSVC TFIDF')
```

### Таблица результатов

```python
pd.DataFrame(wf.results).transpose()[['f1 CV score']]
```

Таким образом две из рассмотренных моделей дали требуемую метрику f1 выше 0.75 и с векторизацией по алгоритму BoW и по алгоритму If-Idf. При этом логистическая регрессия показывает лучший результат. Подберём оптимальные параметры.

### Оценка лучшей модели на тесте

```python
model = make_pipeline(tfidf, LogisticRegression(random_state=31416, n_jobs=-1))
```

```python
param_grid = {
    'logisticregression__solver':['lbfgs', 'liblinear'],
    'logisticregression__C':[1, 5, 10],
    'logisticregression__class_weight':['balanced', None]
}

gs_lr = GridSearchCV(
    model,
    param_grid,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)
```

```python
%time gs_lr.fit(X, y)
None
```

```python
pd.DataFrame(gs_lr.cv_results_)[[
    'param_logisticregression__C', 'param_logisticregression__class_weight',
    'param_logisticregression__solver', 'mean_test_score', 'rank_test_score'
]]
```

Проверим модель на тестовых данных.

```python
wf.add_model(gs_lr.best_estimator_, 'LogisticRegression BOW final', X, y, X_test, y_test, fit=False, report=True)
```

### Выводы по моделям

Две модели из трех на кросс-валидации показали метрику f1 выше требуемой 0.75.

Проведен подбор гиперпараметров для оптимальной модели LogisticRegression с последующей проверкой качества модели на тестовом наборе данных. Получен результат качества f1 выше 0.75.

Модель с высокой точностью определяет нетоксичные комментарии, но пропускает около 25% токсичных.

```python
tokenizer = transformers.BertTokenizer.from_pretrained('unitary/toxic-bert')
None
```

```python
import random
ind_500 = random.sample(list(df.index), 500)
df_500 = df[df.index.isin(ind_500)].reset_index(drop=True)
display(df_500.head(), df_500.tail())
df_500.shape
```

```python
text = list(df_500['text'])
None
```

```python
text
```

```python
%%time
tokenized = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
None
```

```python
tokenized
```

```python
model = transformers.BertModel.from_pretrained('unitary/toxic-bert')
None
```

```python
%%time
batch_size = 50
embeddings = []

for i in notebook.tqdm(range(tokenized['input_ids'].shape[0] // batch_size)):
    
        tokens = tokenized['input_ids'][batch_size*i:batch_size*(i+1)]
        token_types = tokenized['token_type_ids'][batch_size*i:batch_size*(i+1)]
        mask = tokenized['attention_mask'][batch_size*i:batch_size*(i+1)]
        
        batch = {'input_ids' : tokens,
                 'token_type_ids' : token_types, 
                 'attention_mask' : mask}
                
        with torch.no_grad():
            batch_embeddings = model(**batch)
        
        embeddings.append(batch_embeddings[0][:,0,:].numpy())
        
#torch.cuda.empty_cache()
None
```

```python
with open('embeddings.pickle', 'wb') as f:
     pickle.dump(embeddings, f)
None
```

```python
'''with open('embeddings.pickle', 'rb') as f:
    embeddings = pickle.load(f)'''
```

```python
X = np.concatenate(embeddings)
y = df_500['toxic'] #[0:499]
```

```python
X, X_test, y, y_test = train_test_split(X, y, random_state=SEED, stratify=y)
```

```python
model = LogisticRegression(random_state=SEED, n_jobs=-1)
```

```python
param_grid = {
    'solver':['liblinear', 'saga'],
    'C':[0.05, 0.1, 0.15]
}

gs_lr = GridSearchCV(
    model,
    param_grid,
    scoring='f1',
    n_jobs=-1,
    verbose=1
)
```

```python
%time gs_lr.fit(X, y)
None
```

```python
pd.DataFrame(gs_lr.cv_results_)[['param_C', 'param_solver', 'mean_test_score', 'rank_test_score']]
```

```python
wf.add_model(gs_lr.best_estimator_, 'LogisticRegression BERT final', X, y, X_test, y_test, fit=False, report=True)
```

Загружена модель BERT и на её основе получены эмбеддинги комментариев. Ограничения тренажера позволили работать с 500 строками таблицы. Передали эмбеддинги в качестве признаков в логистическую регрессию и получили предсказания. Т.к. модель BERT base была обучена дообучена на схожей задаче, такая связка показала отличный результат - f1 токсичных комментариев на тесте больше 0.88. С увеличением выборки данный показатель будет еще больше.
