### **Описание процесса обработки стали**

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены на крышке ковша.

Сначала происходит десульфурация — из стали выводят серу и корректируют её химический состав добавлением примесей. Затем сталь легируют — добавляют в неё куски сплава из бункера для сыпучих материалов или порошковую проволоку через специальный трайб-аппарат.

Прежде чем в первый раз ввести легирующие добавки, специалисты производят химический анализ стали и измеряют её температуру. Потом температуру на несколько минут повышают, уже после этого добавляют легирующие материалы и продувают сталь инертным газом, чтобы перемешать, а затем снова проводят измерения. Такой цикл повторяется до тех пор, пока не будут достигнуты нужный химический состав стали и оптимальная температура плавки.

Дальше расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).

### **Соберем вместе необходимые функции для работы с данными**

```python
def convert_time_object(time_obj):
    """
    Преобразует объект datetime.time в общее количество секунд от начала суток.

    :param time_obj: Объект datetime.time, который нужно преобразовать.
    :return: Общее количество секунд от начала суток.
    """

    time_str = time_obj.isoformat()
    h, m, s = map(int, time_str.split(':'))
    return h * 3600 + m * 60 + s


def safe_time_conversion(time_obj):
    """
    Безопасно преобразует объект времени в общее количество секунд от начала суток.
    Функция проверяет, является ли объект экземпляром datetime.time, перед тем как произвести расчет.

    :param time_obj: Объект времени, который предполагается преобразовать.
    :return: Общее количество секунд от начала суток, если объект является экземпляром datetime.time; в противном случае 0.
    """

    if isinstance(time_obj, time):
        return time_obj.hour * 3600 + time_obj.minute * 60 + time_obj.second
    else:
        return 0


# Функция для проверки, является ли строка действительным представлением времени
def is_time_format(s):
    try:
        datetime.strptime(s, '%H:%M:%S')
        return True
    except ValueError:
        return False

# Функция для расчета start_time, исключая 'key' и нули
def get_start_time(row):
    values = row.drop('key')  # исключаем 'key'
    non_zero_values = values[values > 0]  # получаем значения больше 0
    if non_zero_values.empty:  # если других значений нет, возвращаем NaN
        return np.nan
    else:
        return non_zero_values.min()  # в противном случае возвращаем минимальное значение


# Функция для вычисления максимального времени в строке, исключая столбец 'key'
def calculate_finish_time(row):
    row_filtered = row.loc[row.index.difference(['key'])]
    if not row_filtered.empty:
        return row_filtered.max()
    else:
        return np.nan  # Возвращаем NaN, если строка пуста
```

```python
# Создаем объект SimpleImputer
imputer = SimpleImputer(strategy='mean')

# Применяем Imputer к тренировочным данным
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Параметры для GridSearch
params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)
grid = GridSearchCV(RandomForestRegressor(random_state=RANDOM_STATE), params, scoring=mae_scorer, cv=5, n_jobs=-1, verbose=1)

# Обучение модели
grid.fit(X_train_imputed, y_train)

# Сохранение лучшего значения ошибки в переменную mae
mae = -grid.best_score_

# Вывод результатов кросс-валидации
print(f"Лучшие параметры: {grid.best_params_}")
print(f"Лучшее значение MAE на кросс-валидации: {mae:.2f}")
```

## Градиентный Бустинг


```python
# Создание модели
gbr_model = GradientBoostingRegressor(random_state=RANDOM_STATE)

# Обучение модели
gbr_model.fit(X_train_imputed, y_train)

# Оценка качества модели с использованием кросс-валидации
scores = cross_val_score(gbr_model, X_train_imputed, y_train, scoring=mae_scorer, cv=5)

# Сохранение среднего значения ошибки в переменную mae_gbr
mae_gbr = -scores.mean()

# Вывод среднего значения ошибки на кросс-валидации
print(f"Средняя абсолютная ошибка (MAE) модели градиентного бустинга на кросс-валидации: {mae_gbr:.2f}")
```

```python
# График важности признаков
importances_gbr = gbr_model.feature_importances_
indices_gbr = np.argsort(importances_gbr)[::-1]

plt.figure(figsize=(12, 6))
plt.title("Важность признаков (Gradient Boosting)")
plt.bar(range(X_train.shape[1]), importances_gbr[indices_gbr], align="center")
plt.xticks(range(X_train.shape[1]), np.array(X_train.columns)[indices_gbr], rotation=90)  # Подписи оси X будут названиями признаков
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

## Обучение модели CATBOOST с использованием гиперпараметров

```python
%%time
# Параметры для тонкой настройки
param_grid = {
    'iterations': [500],
    'depth': [8],
    'learning_rate': [0.05],
    'l2_leaf_reg': [7]
}

# Инициализация модели CatBoost # с использованием GPU
model = CatBoostRegressor(random_seed=RANDOM_STATE, verbose=False) #, task_type='GPU', devices='0:1')

# Инициализация GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Вывод лучших параметров
best_params = grid_search.best_params_
print(f"Лучшие параметры: {best_params}")

# Обучение модели с лучшими параметрами # и использованием GPU
best_model = CatBoostRegressor(**best_params, random_seed=RANDOM_STATE, verbose=False) #, task_type='GPU', devices='0:1')
best_model.fit(X_train, y_train)

# Кросс-валидация с лучшей моделью
mae_catboost = abs(cross_val_score(best_model, X=X_train, y=y_train, cv=5, scoring='neg_mean_absolute_error')).mean()
print(f"MAE (CatBoost): {round(mae_catboost, 2)}")
```

```python
# Получение важности признаков
feature_importances = best_model.get_feature_importance()

# Отображение важности признаков на графике
plt.figure(figsize=(12, 6))
indices = np.argsort(feature_importances)[::-1]
plt.title("Важность признаков для CatBoost")
plt.bar(range(X_train.shape[1]), feature_importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), np.array(X_train.columns)[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

## Нейронная сеть

```python
# Нормализация данных
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# Создание модели
model = Sequential()
input_dim = X_train_scaled.shape[1]

# Добавление слоев
model.add(Dense(128, input_dim=input_dim, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))  # линейная активация для регрессии

# Компиляция модели
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_absolute_error')

# Обучение модели
history = model.fit(X_train_scaled, y_train, validation_data=(X_test_scaled, y_test), epochs=100, batch_size=32, verbose=1)

# Извлечение MAE для тестовых данных из истории обучения
mae_nn = history.history['val_loss'][-1]
print(f"Средняя абсолютная ошибка (MAE) нейросети на тестовых данных: {round(mae_nn, 2)}")
```

```python
# Визуализация истории потерь
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Тренировочные данные')
plt.plot(history.history['val_loss'], label='Валидационные данные')
plt.title('История потерь нейросети')
plt.ylabel('Средняя абсолютная ошибка')
plt.xlabel('Эпоха')
plt.legend()
plt.show()
```

### Оценка R²

```python
# делаем прогнозы на тестовом наборе данных
y_pred = best_model.predict(X_test)

# вычисляем R²
r2 = r2_score(y_test, y_pred)

print(f'R²: {r2}')
```

R² показывает, что дисперсия входящих признаков только приблизительно в половине  случаев объясняет дисперсию целевого признака.

```python
def plot_feature_importances(model, X_train):
    feature_names = X_train.columns
    importances = model.feature_importances_
    model_importances = pd.Series(importances, index=feature_names)

    # Вычисление стандартного отклонения важности признаков
    try:
        std = np.std([tree.feature_importances_ for tree in model.estimators_], axis=0)
    except AttributeError:
        std = None

    fig, ax = plt.subplots(figsize=(10,7))
    model_importances.plot.bar(yerr=std, ax=ax)
    ax.set_title("Feature importances")
    ax.set_ylabel("Mean decrease in impurity")
    fig.tight_layout()
    plt.show()


# Лучшая модель на основе MAE
bestModel = best_model

plot_feature_importances(bestModel, X_train)
```

```python
# Получение важности признаков
feature_importances = best_model.get_feature_importance()

# Получение названий признаков
feature_names = X_train.columns

# Создание DataFrame для хранения названий и важности признаков
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': feature_importances
})

# Сортировка признаков по убыванию важности
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Вывод важности признаков
print(importance_df)
```
