Стационарный временной ряд - это тот, свойства которого не зависят от времени, в которое этот ряд наблюдается. Таким образом, временные ряды с трендами или сезонностью не являются стационарными - тренд и сезонность будут влиять на значение временного ряда в разное время. С другой стороны, серия белого шума является стационарной - не имеет значения, когда вы ее наблюдаете, она должна выглядеть примерно так же в любой момент времени.

Это связано с тем, что циклы не имеют фиксированной длины, поэтому, прежде чем мы будем наблюдать за сериями, мы не можем быть уверены, где будут пики и впадины циклов. То есть статистические свойства ряда, такие как среднее значение, дисперсия и автокорреляция, постоянны во времени.

Проведем проверку стационарности временного ряда. Воспользуемся критерием Дики-Фуллера

<div class="alert alert-info">
<h2> Комментарий студента <a class="tocSkip"> </h2>
H 0 : Временной ряд является нестационарным. Другими словами, он имеет некоторую структуру, зависящую от времени, и не имеет постоянной дисперсии во времени.

H A : временной ряд является стационарным.

Если p-значение из теста меньше некоторого уровня значимости (например, α = 0,05), то мы можем отвергнуть нулевую гипотезу и сделать вывод, что временной ряд является стационарным.
</div>

```python
test = adfuller(data['num_orders'], autolag='AIC')
```

```python
test
```

```python
print(f'adf: {test[0]}')
print(f'p-value: {test[1]}')
print('Критические значения:')

for key, value in test[4].items():
    print('\t%s: %.3f' % (key, value))
    if test[0] < test[4][key]:
        print (f"Нулевая гипотеза отвергнута {key} – Временной ряд стационарен")
    else:
        print (f"Нулевая гипотеза не отвергнута {key} – Временной ряд не стационарен")
```

## Обучение

В датасете всего 1 столбец - количество заказов. Для того, чтобы обучить модели нам надо создать дополнительные признаки. Для этого напишем функцию make_features().

Календарные признаки. Так как в нашей таблице содержатся только данные за пол года, то выделим следующие признаки: месяц, день, день недели - категориальное значение, часы. Кроме того, добавим категориальное значение - время суток.

Предыдущие значения. Cоздадим столбцы с отстающими значениями с максимальным размером отставания в 2 часов.

Cкользящее среднее. Ранее мы выяснили, что пики заказов приходятся на промежутки по 2-4 часа. Поэтому зададим ширину окна в 4 часа. Помимо этого создадим скользящее среднее по дням (24 часа) и за 3 дня, так как ранее мы выявили недельную сезонность.

На вход функцию make_features() будет принимать последовательно датасет, размер максимального отставания, ширину окна для скользящего среднего.

```python
df = data.copy(deep=True)
```

```python
def make_features(df, max_lag, rolling_mean_size):
    #df['month'] = df.index.month
    df['day'] = df.index.day
    df['dayofweek'] = df.index.dayofweek.astype('category')
    df['hour'] = df.index.hour
    
     # добавим время суток
    df.loc[(df.index.hour >= 0) & (df.index.hour < 6), 'times_of_day'] = 'night'
    df.loc[(df.index.hour >= 6) & (df.index.hour < 12), 'times_of_day'] = 'morning'
    df.loc[(df.index.hour >= 12) & (df.index.hour < 18), 'times_of_day'] = 'day'
    df.loc[(df.index.hour >= 18) & (df.index.hour < 24), 'times_of_day'] = 'evening'
    df['times_of_day'] = df['times_of_day'].astype('category')
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(rolling_mean_size).mean()
    df['rolling_mean_2'] = df['num_orders'].shift().rolling(24).mean()#скользящее среднее за сутки
    df['rolling_mean_3'] = df['num_orders'].shift().rolling(72).mean()#скользящее среднее за 3 суток
```

### ExtraTreesRegressor

```python
n_features = X_lr.shape[1]
n_samples = X_lr.shape[0]
```

```python
%%time

param_grid = {
    'max_depth': [None,1,2,3],
    'max_features': [None, 'sqrt', 'auto', 'log2', n_features//2, n_features//3],
    'min_samples_split': [2,3, n_samples//2, n_samples//3, n_samples//5],
    'min_samples_leaf':[2,3,5,  n_samples//2, n_samples//3, n_samples//5]
}

gs_etr = GridSearchCV(
    ExtraTreesRegressor(random_state=SEED),
    param_grid,
    scoring=neg_root_mean_squared_error_scorer,
    n_jobs=-1,
    cv=tscv,
    verbose=5
)

#for max_depth in tqdm(([None,1,2,3]), desc='progress'):
etr = gs_etr.fit(X_lr, y)
```
